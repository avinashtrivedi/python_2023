```python
import pandas
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
dataset = pd.read_excel("D:\OneDrive - NITT\Custom_Download\Online_Retail.xlsx")
```

```python
dataset
```

```python
df = dataset.dropna()
```

```python
df
```

```python
df = df[(df['Quantity'] > 0) & (df['UnitPrice'] > 0)]
```

```python
df
```

```python
#multiplying Quantity and Unitprice 
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']
```

```python
customer_spending = df.groupby("CustomerID")["TotalPrice"].sum()
```

```python
#printing customer_spending
customer_spending.values
```

```python
grouped = df.groupby('CustomerID')['TotalPrice'].agg(['mean', 'median', 'std'])
```

```python
grouped
```

```python
plt.hist(customer_spending.values,bins=50,rwidth=0.7,color='red')
plt.xlabel('Total Amount Spent')
plt.ylabel('Count')
plt.title('Distribution of Total Amount Spent by Customers')
plt.show()
```
