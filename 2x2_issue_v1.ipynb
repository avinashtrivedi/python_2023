```python
def custom_agg(x):
    return pd.Series({
        
    'unique_BUY_cur': x.BUY_CCY_ID.nunique(), 
    'count_BUY_cur': dict(x.BUY_CCY_ID.value_counts()),
    'unique_SELL_cur': x.SELL_CCY_ID.nunique(),
    'count_SELL_cur': dict(x.SELL_CCY_ID.value_counts()),
    
    })
```

```python
df.groupby(['COMPANY_ID','LEGAL_ENTITY_ID','VALUE_DATE']).apply(custom_agg)
```
