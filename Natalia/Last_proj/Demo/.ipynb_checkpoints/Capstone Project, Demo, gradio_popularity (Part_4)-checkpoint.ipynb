The PassthroughTransformer class takes from the BaseEstimator class and defines a transformer that performs a passthrough of the input data. The fit method stores the column names of the input data.

```python
class PassthroughTransformer(BaseEstimator):
    
    # Save the column names of the input dataframe
    def fit(self, X, y = None):
        self.cols = X.columns
        return self

 # The transform method takes the input data and creates new binary features 
 #    for each genre present in the 'genres' column.


    def transform(self, X, y = None):
        # Create a copy of the input dataframe
        X_ = X.copy()
        all_genre = {'Action',
                     'Adventure',
                     'Animation',
                     'Comedy',
                     'Crime',
                     'Documentary',
                     'Drama',
                     'Family',
                     'Fantasy',
                     'Foreign',
                     'History',
                     'Horror',
                     'Music',
                     'Mystery',
                     'Romance',
                     'Science Fiction',
                     'TV Movie',
                     'Thriller',
                     'War',
                     'Western'}

# It iterates over all genres listed above using a for loop and applies a lambda function
 # to assign a value of 1 if the genre is present in a movie, and 0 otherwise. '''
        for gen in tqdm(all_genre):
            X_[gen] = X_['genres'].apply(lambda x: 1 if gen in x else 0)

       # Remove the 'genres' column  
        X_.drop('genres',axis=1,inplace=True)
        return X_

        # Return the saved column names of the input dataframe
    def get_feature_names(self):
        return self.cols


# https://stackoverflow.com/questions/53382322/adding-get-feature-names-to-columntransformer-pipeline
```
