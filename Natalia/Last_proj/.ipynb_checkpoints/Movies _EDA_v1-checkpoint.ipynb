```python
import requests
```

The get_data() function retrieves movie data from The Movie Database API by sending a GET request with a specified URL and headers. The input is an IMDb ID, and the output is a response object containing the movie data. The function also supports retrieving data by movie ID using an optional flag parameter. If an error occurs, the function prints an error message.

```python
# sometimes websites stop you from extracting the data for some reason. It can be due to some authentication errors.

def get_data(ex_id,flag=False):
    needed_headers = {'User-Agent': "Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/92.0.4515.131 Safari/537.36"}

#     url = "https://api.themoviedb.org/3/movie/popular?api_key=81b49b634e27bb42bfe9fd2a462146c0&page=1"
    if flag:
        url = f"https://api.themoviedb.org/3/movie/{ex_id}?api_key=81b49b634e27bb42bfe9fd2a462146c0"
    else:
        url = f"https://api.themoviedb.org/3/find/{ex_id}?api_key=81b49b634e27bb42bfe9fd2a462146c0&language=en-US&external_source=imdb_id"
    response = requests.get(url, headers = needed_headers)
    if response.status_code !=200:
        print('Error',ex_id)
    return response
```

The get_json() function retrieves movie data from The Movie Database API using the get_data() function and an IMDb ID as input. The function then extracts the movie ID from the response object and sends a second GET request to the API using the get_data() function. The output of the function is a dictionary object containing the movie data, or an empty dictionary if an error occurs.

```python
def get_json(imdb_id):
    response = get_data(imdb_id,True)
    try:
        api_id = response.json()['movie_results'][0]['id']
        response = get_data(api_id)
        data = response.json()
        if data['revenue']!=0:
            return data
        else:
            return {}
    except:
        return {}
```

```python
def plot_items_per_year(df, date_col):

    # Convert the date column to a datetime data type
    df[date_col] = pd.to_datetime(df[date_col], format='%Y-%m-%d')

    # Extract the year from the date column
    df['year'] = df[date_col].dt.year

    # Count the number of items released per year
    year_counts = df['year'].value_counts().sort_index()

    # Create a line plot of the number of items released per year
    plt.plot(year_counts.index, year_counts.values)

    # Set the plot title and axis labels
    plt.title('Number of Items Released per Year')
    plt.xlabel('Year')
    plt.ylabel('Number of Items')

    # Show the plot
    plt.show()
```

```python
import plotly.express as px

# Top 10 most popular movies
top_popularity = tmdb_5000_mov.sort_values('popularity', 
                                           ascending=False).head(10)
# Round the popularity values to 2 decimal places
top_popularity['popularity'] = top_popularity['popularity'].round(0) 

# Create a bar chart of the top 10 most popular movies
fig = px.bar(top_popularity, x='title', y='popularity', 
             color='title', text='popularity', color_continuous_scale=px.colors.sequential.Teal,
             width=1000,
             height=600)
fig.update_layout(title='Top 10 Most Popular Movies', 
                  xaxis_title='Title', yaxis_title='Popularity', plot_bgcolor='rgba(0,0,0,0)')

fig.show()

```

**Pularity** IMDbPro uses proprietary algorithms that take into account several measures of popularity for people, titles and companies. The primary measure is who and what people are looking at on IMDb. The rankings are updated on a weekly basis, typically by the end of Monday.

**Vote Count** represents the number of votes that a movie has received on the TMDB platform, which can be a useful indicator of a movie's overall popularity and the size of its fanbase. However, the number of votes a movie receives is influenced by various factors such as marketing, distribution, and the number of people who have access to the platform. Moreover, it is possible for a movie to have a large number of votes but low average ratings or reviews, which may affect its overall success.

##### Adding a new column: profit

```python
# Dropping rows where budget or revenue is zero
tmdb_5000_mov = tmdb_5000_mov.drop(
    tmdb_5000_mov[(tmdb_5000_mov['budget'] == 0) |
                  (tmdb_5000_mov['revenue'] == 0)].index)

# Calculating profit as revenue - budget
tmdb_5000_mov['profit'] = tmdb_5000_mov['revenue'] - tmdb_5000_mov['budget']

# Creating a scatter plot of profit vs. popularity
fig = px.scatter(tmdb_5000_mov, x='profit', y='popularity', hover_name='title')

# Calculating correlation coefficient
corr = tmdb_5000_mov['profit'].corr(tmdb_5000_mov['popularity'])

# Adding correlation coefficient as text annotation
fig.add_annotation(x=0.9, y=0.1, showarrow=False,
                   text=f'Correlation: {corr:.2f}', font=dict(size=16))

# Updating layout with title and axis labels
fig.update_layout(title='Profit vs. Popularity', xaxis_title='Profit', yaxis_title='Popularity')

# Showing the plot
fig.show()

```
