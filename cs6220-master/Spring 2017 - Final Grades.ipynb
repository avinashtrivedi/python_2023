```python
%pylab inline

import matplotlib.pyplot as plt
matplotlib.style.use('ggplot')

import pandas as pd
```

```python
df = pd.read_csv('detailed_scores.csv')
```

## Histogram with overall grade distribution

```python
hist = df.TOTAL.hist(bins=10)
plt.title('Midterm Grade Distribution')
plt.ylabel('Number of Exams')
plt.xlabel('Grade')
```

## Summary Stats (mean, std, min, max, percentiles)

```python
summary_stats = df.TOTAL.describe()[1:] 
for i,x in zip(summary_stats.index,summary_stats):
    print(i, round(x,1))
```

## Histogram of scores per question

```python
hist = df.ix[:,:11].hist(figsize=(10,10), bins=range(11), color='brown')
```

## Mean scores for each question

```python
question_means = df.mean()[:11]
ax = question_means.plot(rot=90, grid=True, color='r')
ax.set_xticks(range(11))
ax.set_yticks(range(11))
ax.lines[-1].set_linewidth(3)
for i,x in enumerate(question_means):
   ax.text(i-.25, x+.6, str(round(x,1)), color='b', weight='bold')
ax.set_xticklabels(df.columns[:11].tolist())
ax.set_ylim(-1, 11)
ax.set_xlim(-1, 11)
ax.set_title('Average scores per question');
```
