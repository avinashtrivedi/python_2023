#### 3.b. Implement a Euclidean distance function  
In order to find the nearest neighbor, a function that computes the Euclidean distance between points in arbitrary dimensions is required. An easy implementation of the KNN algorithm is to loop through all training points and calculate the distance between the testing point and a single training point at each iteration of the loop.

However, the loop implementation is often slow when we have a large dataset. A better way to do this is to calculate the distances between the testing point and all training points using a **vectorized implementation**.  
For full credit, implement the `euclidean_distance` function below __without using a loop__!

```python
def euclidean_distance(A, b):
    '''
    A function that returns the Euclidean distances
    between vector b and each of the row vectors in matrix A

    :A: np.array with shape [n, m]
    :b: np.array with shape [m, ]

    :return: an array with distances between b and each row of A [n, ]
    '''
    ## YOUR CODE HERE
    
    # Calculate the squared differences element-wise
    squared_diff = (A - b)**2
    
    # Sum the squared differences along the second axis (axis=1) to get the squared distances
    squared_distances = np.sum(squared_diff, axis=1)
    
    # Take the square root to get the Euclidean distances
    distances = np.sqrt(squared_distances)
    return distances
```

#### 3.c. Implement the K-nearest neighbors algorithm.  
The two functions below make up a skeleton of the KNN algorithm, which you will complete.

The first, `compute_neighbors_and_classify()`:
1. Computes the neighbors given `k` (the number of neighbors)
2. Finds the `k` neighbors with the smallest Euclidean distance to the test point (using `euclidean_distance()` above)
3. Returns the most common classification label among those `k` neighbors.

The second, `compute_accuracy_on_dataset()`:
1. Gets the KNN classification predictions for each test point using `compute_neighbors_and_classify()`.
2. Compares the predictions to the real labels.
3. Returns the accuracy score (how many predictions the model got right divided by total number of samples).

To receive full credit, You __may not use a loop__ in the `compute_neighbors_and_classify()` function. If you feel like it's easier to calculate the distance within the `compute_neighbors_and_classify()` function instead of implementing the distance function, you may do so.

```python
from scipy.stats import mode
import numpy as np

def compute_neighbors_and_classify(X_train, y_train, test_point, k):
    '''
    Implementation of the KNN algorithm

    :return: predicted classification of test_point
    '''
    # calculate the distance between test_point and the points in X_train
    ### Add code here
    distances = np.sqrt(np.sum((X_train - test_point) ** 2, axis=1))
    
    # Now that we have all the distances, we need to return the labels with the k smallest distances
    # Hint: you might want to use np.argsort somewhere
    ### Add code here
    nearest_indices = np.argsort(distances)[:k]

    # Now, pull out the labels associated with the first k neighbors
    ### Add code here
    k_labels = y_train[nearest_indices]
    
    # Get the label that appears the most times in k_labels using mode(). This is our classification.
    ### Add code here
    classification = mode(k_labels).mode

    return classification

def compute_accuracy_on_dataset(X_train, y_train, data, labels, k):
    '''
    A function that computes the accuracy of KNN on a (test) dataset
    '''
    accuracy_numerator = 0

    # loop over the dataset we'd like to get the accuracy on
    for i, point in enumerate(data):
        # compute the neighbors and classification for this point with the training data
        ### Add code here
        distances = np.sqrt(np.sum((X_train[:, np.newaxis, :] - data) ** 2, axis=2))
        nearest_indices = np.argsort(distances, axis=0)[:k]
        k_labels = y_train[nearest_indices]
        
        # compare this classification to the real value in labels
        # if the model got it correct, add to a running sum of correct predictions.
        ### Add code here
        classifications = mode(k_labels, axis=0).mode.flatten()
        correct_predictions = np.sum(classifications == labels)
    # Compute the accuracy: divide the count of correct predictions by the number of samples in the dataset
    ### Add code here
    accuracy = correct_predictions / len(labels)

    return accuracy
```

#### 3.d. Run KNN with `k = [1, 50, 150, 300]`, and report the accuracy of each `k` on the test set.

```python
k = [1, 50, 150, 300]
### Add code here
for i in k:
    print(f'Accuracy on k={i}',compute_accuracy_on_dataset(X_train, y_train, X_test,y_test, i))
```

#### 3.e. Parameter Inspection
Run the following code block which plots the decision boundary and the entire dataset when `k = [1, 50, 150, 300]`, respectively. Combining with the accuracy on testing set reported from the previous question:
  

1. Explain the value of `k` that you would choose that fit the given dataset the best
2. Explain why the other 3 values for `k` were _not_ chosen.

```python
from matplotlib.colors import ListedColormap
from tqdm.notebook import tqdm

def generate_decision_boundary(X_train, y_train, xx, yy, k):

    pred = np.zeros_like(xx)
    text = "KNN with k=" + str(k)
    for i in tqdm(np.arange(xx.shape[0]), desc=text):
        for j in np.arange(xx.shape[1]):
            pred[i,j] = compute_neighbors_and_classify(X_train, y_train, [xx[i,j], yy[i,j]], k)
    return pred

cmap_bold = ListedColormap(['#d7191c', '#2c7bb6'])
cmap_light = ListedColormap(['#fdae61', '#abd9e9'])

x_min, x_max = X[:,0].min()-0.1, X[:,0].max()+0.1
y_min, y_max = X[:,1].min()-0.1, X[:,1].max()+0.1
h = 0.03

xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                    np.arange(y_min, y_max, h))


fig, axs = plt.subplots(1, len(k), figsize=(20, 5))
for i, knn in enumerate(k):
    pred = generate_decision_boundary(X_train, y_train, xx, yy, int(knn))
    axs[i].pcolormesh(xx, yy, pred,cmap=cmap_light)
    axs[i].scatter(X[:,0],X[:,1], c= y, cmap=cmap_bold, s= 10)
    axs[i].set_title('k = %s' %(knn))
```

KNN has a bias-variance trade-off. A smaller K (e.g., K=1) leads to a more complex model with lower bias but higher variance. This can make the model sensitive to noise in the data and result in overfitting. Conversely, a larger K (e.g., K=50) leads to a smoother decision boundary, which reduces variance but may introduce bias. K=50 maintain a balance between bias and variance that fits the dataset well, Similarly for other value of K bias-variance trade-off is not favourable

### 4. Comparing Decision Trees and K-Nearest Neighbors on Raw Wine Data (15 points)
We're now going to compare the performance of our KNN algorithm and the decision tree algorithm on a new dataset.

The dataset we'll be using is the popular wine dataset. The data is the results of a chemical analysis of wines grown in the same region in Italy by three different cultivators. There are 13 different measurements (features) taken for different constituents found in the three types of wine.

#### Import and view the wine dataset

```python
from sklearn.datasets import load_wine
import pandas as pd

# Import the data into a pandas dataframe
wine = load_wine()
wine_df = pd.DataFrame(wine.data, columns = wine.feature_names)

# Split data into features and labels
X_w = wine.data
y_w = wine.target

# View the raw data
wine_df.head()
```

#### 4.a. Split the data into train, test and validation sets.

Use a 80/20 split. Make sure to set `random_state=42` to shuffle the dataset in a consistent manner.

```python
# Split the data into training and test
# Make sure to set random_state=42!
### Add code here
X_train, X_test, y_train, y_test = train_test_split(X_w, y_w, test_size=0.20, random_state=42)
```

#### 4.b.  KNN: Use `sklearn`'s built-in KNN on this dataset and report the accuracy.
You can find the documentation for instantiating and fitting `sklearn`'s `KNeighborsClassifier` [here](https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsClassifier.html).

You can find the documentation for using `metrics.accuracy_score` to compute the accuracy given predictions and true labels [here](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.accuracy_score.html).


```python
from sklearn import neighbors
from sklearn import metrics


# Instantiate the KNN with k=4
### Add code here
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=4)

# Fit the KNN model with training set.
### Add code here

neigh.fit(X_train,y_train)

# Use the KNN model to predict on the testing set.
### Add code here
pred = neigh.predict(X_test)

# Use metrics.accuracy_score to get the accuracy of this decision tree on the testing set.
### Add code here

knn_acc = metrics.accuracy_score(y_test,pred)
print("KNN accuracy on test set:", knn_acc)
```
