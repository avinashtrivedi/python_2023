# Categorical variables to binary variables conversion

To convert categorical attributes to binary variables, we can use a technique called one-hot encoding. This involves creating a new binary variable for each category of the attribute, with a value of 1 indicating that the sample belongs to that category and a value of 0 indicating that it does not.

For example, in the case of the Fuel Type attribute,options are "Diesel","CNG" and "Petrol", we would create Three binary variables: Fuel_Type_Petrol,Fuel_Type_CNG and Fuel_Type_Diesel. A sample with a Fuel Type of "Diesel" would have a value of 1 for the Fuel_Type_Diesel variable and a value of 0 for the Fuel_Type_CNG and Fuel_Type_Petrol variable. Similarly, for other Fuel types.

The same process applies to the Color attribute.

```python
# perform categorical variable into dummy variables.
df = pd.get_dummies(df,columns=['Fuel_Type', 'Color'])
df
```

```python
# number of column is increased after performing get_dummies 
df.shape
```

# Weight normalization

```python
# weight before normalization
df['Weight']
```

```python
# perform mean normalization
# (x-x_mean)/x_std
df['Weight'] = (df.Weight - df.Weight.mean())/df.Weight.std()
```
