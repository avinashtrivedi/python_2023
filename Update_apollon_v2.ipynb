```python
df = pd.read_csv(r"D:\OneDrive - NITT\Custom_Download\wochenumsatz (1).csv",sep=';',encoding='unicode_escape')
```

```python
df['UVKI'] = df['UVKI'].str.replace("'","")
df['UVKI'] = df['UVKI'].astype("float")
```

```python
ddf = df.groupby(['FIL','ZR'])['UVKI'].sum()
```

```python
ddf = ddf.reset_index()
```

```python
th = ddf['ZR'].max() - ddf['ZR'].min() + 1
```

```python
month_min = ddf['ZR'].min()
```

```python
weeks = list(range(month_min,month_min+th))
```

```python
weeks
```

```python
for s_no in sorted(set(ddf['FIL'])):
    l = len(ddf[ddf['FIL']==s_no])
    if l<th:
#         print(s_no)
        for w in weeks:
            if w not in list(ddf[ddf['FIL']==s_no]['ZR']):
                dummy_data = [s_no,w,0]           
                ddf.loc[len(ddf)] = dummy_data  # adding a row
```

```python
ddf = ddf.sort_values(['FIL','ZR'])
```

```python
ddf
```

```python
ddf = ddf.pivot_table(index=['ZR'], columns=['FIL'], values='UVKI', fill_value=0, aggfunc='sum')
```

```python
ddf = ddf.reset_index()
```

```python
ddf = ddf.rename(columns={"ZR": "woche"})
```

```python
ddf
```

```python
ddf.rename_axis(None, axis=1,inplace=True)
```

```python
ddf["Total"] = ddf[list(ddf)[1:]].sum(axis=1)
```

```python
ddf.loc[len(ddf)] = [float("NaN")]*ddf.shape[1]
```

```python
ddf
```

```python
ddf.loc[len(ddf)] = ['Total']+list(ddf.sum())[1:]
```
