```python
data = df_movies[['revenue','runtime','vote_average','vote_count','original_language','budget','popularity']]
```

```python
data
```

```python
# modify text into numeric i.e categorical to numeric column
```

```python
data['original_language'].unique()
```

```python
data = pd.get_dummies(data)
```

```python
data
```

```python
data.dtypes
```

```python
data.dropna(inplace=True)
```

```python
X = data.drop('popularity',axis=1)
```

```python
y = data['popularity']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
X_train
```

```python
# Linear regression

from sklearn.linear_model import LinearRegression
```

```python
reg_model = LinearRegression()
```

```python
reg_model.fit(X_train,y_train) 
```

```python
y_pred = reg_model.predict(X_test)
```

```python
y_pred # predicted popularity
```

```python
y_test # actual test popularity
```

```python
pd.DataFrame({'y_actual':y_test,'y_pred':y_pred})
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mse = mean_squared_error(y_test,y_pred)  #True returns MSE value, if False returns RMSE value.
mse
```

```python
rmse = mean_squared_error(y_test, y_pred, squared=False)
rmse
```
