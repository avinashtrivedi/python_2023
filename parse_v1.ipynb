```python
with open("D:\OneDrive - NITT\Custom_Download\out_query.txt") as fp:
    data = fp.readlines()
```

```python
condition_list = ["CREATE PROCEDURE","CALL","call",'FROM','INNER JOIN','LEFT JOIN']
```

```python
stmt_list = []
for indx,line in enumerate(data):
    
    temp = line[:]
    temp = temp.strip()
    
    if temp:
        if temp.startswith("--"):
    #         print('True')
            temp = temp.replace("--",'')

        for cond in condition_list:
            if temp[0] == '"':
#                 print(line)
                stmt_list.append(line)
                break

            if temp.startswith(cond):
#                 print(line)
                stmt_list.append(line)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame([],columns=['col1','col2','col3'])
```

```python
for line in stmt_list:
    if line.count('.') == 2:
        cols = [line.strip().split('.')[0].split()[-1],line.strip().split('.')[-2],line.strip().split('.')[-1]]
        df.loc[len(df)] = cols
        
#         print(line)
```

```python
df
```

```python
def get_parameters(x):
    start = x.find("(") + 1
    end = x.find(")")
    return x[start:end].split(',')
```

```python
df['col4'] = df['col3'].apply(get_parameters)
```
