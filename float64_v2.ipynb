```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("D:\OneDrive - NITT\Custom_Download\degrees-that-pay-back.csv")
```

```python
def clean(x):
    if isinstance(x,str):
        for i in ('$',','):
            x = x.replace(i,'')
    return x
```

```python
cols = ['Starting Median Salary',
 'Mid-Career Median Salary',
 'Percent change from Starting to Mid-Career Salary',
 'Mid-Career 10th Percentile Salary',
 'Mid-Career 25th Percentile Salary',
 'Mid-Career 75th Percentile Salary',
 'Mid-Career 90th Percentile Salary']
```

```python
for col in cols:
    df[col] = df[col].apply(lambda x:clean(x))
    df[col] = df[col].astype('float64')
```

# New code Added

```python
# get dataframe containing all float columns
df_float = df.select_dtypes('float64')
```

```python
x = list(df['Undergraduate Major'])
plt.figure(figsize=(15,5))
for i in df:
    if i!='Undergraduate Major':
        plt.plot(df[i],label=str(i))
        plt.legend()
plt.xticks(range(50), x, rotation='vertical')
plt.show()
```
