```python
import itertools
import numpy as np

def find_triclusters(*matrices):
    """
    Find all possible triclusters in N binary matrices.

    Args:
        *matrices: N binary matrices.

    Returns:
        list: List of triclusters, where each tricluster is represented as a tuple
              containing the indices of the rows, columns, and matrices that form it.
    """
    num_matrices = len(matrices)
    num_rows, num_cols = matrices[0].shape

    # Initialize list to store all possible triclusters
    triclusters = []

    # Generate all possible combinations of rows, columns, and matrices
    # to form triclusters
    for rows, cols, mats in itertools.product(range(num_rows), range(num_cols), range(num_matrices)):
        # Check if the value at the current position in all matrices is 1
        if all(matrices[i][rows, cols] == 1 for i in range(num_matrices)):
            # Add the tricluster to the list
            triclusters.append((rows, cols, mats))

    # Remove duplicate triclusters
    triclusters = list(set(triclusters))

    # Sort triclusters by row indices, then by column indices, then by matrix indices
    triclusters.sort(key=lambda x: (x[0], x[1], x[2]))

    return triclusters
```
