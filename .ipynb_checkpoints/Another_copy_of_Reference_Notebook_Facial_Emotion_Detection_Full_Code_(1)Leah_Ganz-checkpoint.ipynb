## **Model Building**

**Think About It:**
* Are Convolutional Neural Networks the right approach? Should we have gone with Artificial Neural Networks instead?

CNN's are a good approach here because CNN's have filters that can detect edges, lines, curves, etc, and that is exactly what facial expressions are composed: of lines and curves - lips curving up or down and eyebrows raised. So it will be useful to employ CNN with these filters which can detect these features.

* What are the advantages of CNNs over ANNs and are they applicable here?

CNN's have an advantage of the convolutional layers which sweep across the image using filters which are primed to detect edges and curves and more features as the network becomes deeper. CNN filters are feature and pattern detectors-  and that is what we want to do here -  detect patterns of facial features that comprise different emotion classes.

### **Creating the Base Neural Network**

```python
#setting the parameters for first CNN model based on grayscale
model = Sequential()

#1st cnn block
#first convolutional block with 64 filters, 2x2 kernel size/filter size, with padding to add layer of zeros around image to protect it from shrinking in size after the convolutions, input shape - grayscale with 1 channel
model.add(Conv2D(64,(2,2),padding = 'same', input_shape = (48,48,1)))
#batch normalization layer
model.add(BatchNormalization())
#leaky relu is an activation function
model.add(LeakyReLU(0.1))
#max pooling takes the most significant features of an image
model.add(MaxPooling2D(pool_size = (2,2)))
#Adding a dropout layer - which some neurons are randomly shut off (by zeroing the values). this is a regularization technique to help with overfitting in the model and help model generalize better.
model.add(Dropout(0.2))

#adding second convolutional block following same pattern as above, this block has a 128 filters
#2nd cnn block
model.add(Conv2D(128,(2,2),padding = 'same'))
model.add(BatchNormalization())
model.add(LeakyReLU(0.1))
model.add(MaxPooling2D(pool_size = (2,2)))
model.add(Dropout(0.2))

#adding a third convolutional block following same pattern as above, this block has a 256 filters

#3rd cnn block
model.add(Conv2D(256,(2,2),padding = 'same'))
model.add(BatchNormalization())
model.add(LeakyReLU(0.1))
model.add(MaxPooling2D(pool_size = (2,2)))
model.add(Dropout(0.2))

#adding a flatten layer
#flatten layer
model.add(layers.Flatten())

#adding first dense/connected layer with 512 neurons
#hidden layer
model.add(Dense(512, activation ="relu"))
model.add(BatchNormalization())
model.add(Dropout(0.5))

#adding the output layer with the softmax activation
#output layer with 4 classes using softmax activation function
model.add(Dense(4, activation='softmax'))

#defining the name of the model
Model_gry1 = model
```

```python
#printing summary of model 1 in grayscale
Model_gry1.summary()
```

### **Compiling and Training the Model**

```python
#compiling the model with the optimizer Adam and learning rate of 0.001, using the loss function of cross entropy - used for categorical output variables
model.compile(loss='categorical_crossentropy', metrics=['accuracy'])
tf.keras.optimizers.Adam(
    learning_rate=0.001,
        name="Adam",)
```

```python

#setting parameters for model_checkpoint. the model will check the validation accuracy rates after each epoch and save the best model.
model_checkpoint = ModelCheckpoint("./modelgry1.h5", monitor='val_accuracy', verbose=1, save_best_only=True, mode='max')

# setting parameters for early stopping. The model will stop the training after the data passes through 5 epochs and the validation accuracy does not improve during that time.
early_stopping = EarlyStopping(monitor = 'val_loss',
                          min_delta = 0,
                          patience = 5,
                          verbose = 1,
                          restore_best_weights = True
                          )
# setting parameter for reducing Learning rate - reduces the learning rate when validation loss or accuracy stops improving
reduce_learningrate = ReduceLROnPlateau(monitor = 'val_loss',
                              factor = 0.2,
                              patience = 5,
                              verbose = 1,
                              min_delta = 0.0001)

callbacks_list = [early_stopping, model_checkpoint, reduce_learningrate]
# the data will run through the model 20 times.
epochs = 20
```

```python
#fitting the model to the validation set

history_modelgry1 = model.fit_generator(generator=train_set, epochs=20,
                              validation_data=validation_set,
                              callbacks=[model_checkpoint, early_stopping, reduce_learningrate],
                              verbose=1)
```

```python
#saving the model in Keras
model.save("Model_gry1.keras")
```

```python
#plotting the graphs for the accuracy and loss respectively of the model on the training vs. the validation data and see how well it performed.

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 7))
title = fig.suptitle('CNN Model 1 Grayscale Validation Accuracies and Loss')

epoch_list = list(range(0,20))
ax1.plot(epoch_list,history_modelgry1.history['accuracy'], label='Train Accuracy')
ax1.plot(epoch_list,history_modelgry1.history['val_accuracy'], label='Validation Accuracy')
ax1.set_xticks(np.arange(0, 20))
ax1.set_ylabel('Accuracy Value')
ax1.set_xlabel('Epoch')
ax1.set_title('Accuracy')
l1 = ax1.legend(loc="best")

ax2.plot(epoch_list, history_modelgry1.history['loss'], label='Train Loss')
ax2.plot(epoch_list, history_modelgry1.history['val_loss'], label='Validation Loss')
ax2.set_xticks(np.arange(0, 20))
ax2.set_ylabel('Loss Value')
ax2.set_xlabel('Epoch')
ax2.set_title('Loss')
l2 = ax2.legend(loc="best")
```

```python
#setting the parameters for second CNN model based on grayscale- trying out a larger architecture, by adding a 4th convolutional block and also added a second dense layer
model = Sequential()

#1st cnn block
#first convolutional block with 64 filters, 2x2 kernel size/filter size, with padding to add layer of zeros around image to protect it from shrinking in size after the convolutions, input shape - grayscale with 1 channel, relu activation
model.add(Conv2D(64,(2,2),padding = 'same', input_shape = (48,48,1)))
model.add(BatchNormalization())
#leaky relu is an activation function
model.add(LeakyReLU(0.1))
#max pooling takes the most significant features of an image
model.add(MaxPooling2D(pool_size = (2,2)))
#Adding a dropout layer - which some neurons are randomly shut off (by zeroing the values). this is a regularization technique to help with overfitting in the model and help model generalize better.
model.add(Dropout(0.2))

#adding second convolutional block following same pattern as above, this block has a 128 filters
#2nd cnn block
model.add(Conv2D(128,(2,2),padding = 'same'))
model.add(BatchNormalization())
model.add(LeakyReLU(0.1))
model.add(MaxPooling2D(pool_size = (2,2)))
model.add(Dropout(0.2))

#adding a third convolutional block following same pattern as above, this block has a 256 filters

#3rd cnn block
model.add(Conv2D(256,(2,2),padding = 'same'))
model.add(BatchNormalization())
model.add(LeakyReLU(0.1))
model.add(MaxPooling2D(pool_size = (2,2)))
model.add(Dropout(0.2))

#adding a 4th convolutional block with 512 filters
#4th cnn block
model.add(Conv2D(512,(2,2),padding = 'same'))
model.add(BatchNormalization())
model.add(LeakyReLU(0.1))
model.add(MaxPooling2D(pool_size = (2,2)))
model.add(Dropout(0.2))

#adding a flatten layer
#flatten layer
model.add(layers.Flatten())

#adding first dense/connected layer with 256 neurons
#hidden layer
model.add(Dense(256, activation ="relu"))
model.add(BatchNormalization())
model.add(Dropout(0.5))


#adding second dense/connected layer with 512 neurons
#hidden layer
model.add(Dense(512, activation ="relu"))
model.add(BatchNormalization())
model.add(Dropout(0.5))

#adding the output layer with the softmax activation
#output layer with 4 classes using softmax activation function
model.add(Dense(4, activation='softmax'))

#defining the name of the model
Model_gry2 = model
```

### **Model Building**

- Try building a layer with 5 Convolutional Blocks and see if performance increases.

```python
#setting the parameters for second CNN model based on grayscale- added a 4th convolutional block and also added a second dense layer
model = Sequential()

#1st cnn block
#first convolutional block with 64 filters, 2x2 kernel size/filter size, with padding to add layer of zeros around image to protect it from shrinking in size after the convolutions, input shape - grayscale with 1 channel, relu activation
model.add(Conv2D(64,(2,2),padding = 'same', input_shape = (48,48,1)))
model.add(BatchNormalization())
#leaky relu is an activation function
model.add(LeakyReLU(0.1))
#max pooling takes the most significant features of an image
model.add(MaxPooling2D(pool_size = (2,2)))
#Adding a dropout layer - which some neurons are randomly shut off (by zeroing the values). this is a regularization technique to help with overfitting in the model and help model generalize better.
model.add(Dropout(0.3))

#adding second convolutional block following same pattern as above, this block has a 128 filters
#2nd cnn block
model.add(Conv2D(128,(2,2),padding = 'same'))
model.add(BatchNormalization())
model.add(LeakyReLU(0.1))
model.add(MaxPooling2D(pool_size = (2,2)))
model.add(Dropout(0.3))

#adding a third convolutional block following same pattern as above, this block has a 256 filters

#3rd cnn block
model.add(Conv2D(256,(2,2),padding = 'same'))
model.add(BatchNormalization())
model.add(LeakyReLU(0.1))
model.add(MaxPooling2D(pool_size = (2,2)))
model.add(Dropout(0.3))

#adding a 4th convolutional block
#4th cnn block
model.add(Conv2D(512,(2,2),padding = 'same'))
model.add(BatchNormalization())
model.add(LeakyReLU(0.1))
model.add(MaxPooling2D(pool_size = (2,2)))
model.add(Dropout(0.3))

#adding a 5th convolutional block
#4th cnn block
model.add(Conv2D(1024,(2,2),padding = 'same'))
model.add(BatchNormalization())
model.add(LeakyReLU(0.1))
model.add(MaxPooling2D(pool_size = (2,2)))
model.add(Dropout(0.3))

#adding a flatten layer
#flatten layer
model.add(layers.Flatten())

#adding first dense/connected layer with 512 neurons
#hidden layer
model.add(Dense(256, activation ="relu"))
model.add(BatchNormalization())
model.add(Dropout(0.5))


#adding second dense/connected layer with 512 neurons
#hidden layer
model.add(Dense(512, activation ="relu"))
model.add(BatchNormalization())
model.add(Dropout(0.5))

#adding the output layer with the softmax activation
#output layer with 4 classes using softmax activation function
model.add(Dense(4, activation='softmax'))

#defining the name of the model
Model_gry3 = model
```
