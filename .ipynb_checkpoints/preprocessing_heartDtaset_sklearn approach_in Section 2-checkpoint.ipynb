```python
#https://www.kaggle.com/fedesoriano/heart-failure-prediction
import pandas as pd
df = pd.read_csv('U:\Courses\Slides\CMP466\codes\Preprocessing\heart.csv')

print(df.shape)
df.describe()
```

```python
y = df['HeartDisease'].to_numpy() #Extract the labels and put them in y. It has to be an array (not dataframe)

y 
```

```python
df = df.drop(columns=['HeartDisease'])
```

```python
#Applying ordinal/integer encoding on one variable

from sklearn.preprocessing import OrdinalEncoder

ord_enc = OrdinalEncoder()

df["ChestPainType_code"] = ord_enc.fit_transform(df[["ChestPainType"]])
#df[["ChestPainType", "ChestPainType_code"]].head(10)
df.head(10)
```

```python
ord_enc.categories_
```

```python
#Applying ordinal/integer encoding for multiple variables:

df[["Sex_code", "RestingECG_code"]] = ord_enc.fit_transform(df[["Sex", "RestingECG"]])

#The returned transformed matrix will replace the original columns, as follows:
#df[["Sex", "RestingECG"]] = ord_enc.fit_transform(df[["Sex", "RestingECG"]])

#df[["ChestPainType", "ChestPainType_code"]].head(10)
df.head(10)
```

```python
ord_enc.categories_
```

```python
#Apply oneHot encoding
from sklearn.preprocessing import OneHotEncoder

onehot_enc = OneHotEncoder()

onehot_results = onehot_enc.fit_transform(df[["ST_Slope"]])


#pd.DataFrame(onehot_results.toarray(), columns = onehot_enc.categories_).head()
```

```python
onehot_results
```

```python
onehot_enc.categories_
```

```python
#To join this data back to the original dataframe.
df = df.join(pd.DataFrame(onehot_results.toarray(), columns = onehot_enc.categories_))
```

```python
df.head()
```

```python
# Pass a list of column names to the columns parameter to drop multiple columns
df = df.drop(columns=['ChestPainType', 'Sex','RestingECG'])
```

```python
df = df.drop(columns=['ST_Slope']) # to drop a single column

# Pass a list of column numbers to the columns parameter to drop multiple columns
#df.drop(df.columns[[1, 3, 7]], axis=1) #to drop columns 1,3 and 7, axis =1 means accross columns, while axis =0 accross rows
```

```python
#how to encode ExerciseAngina ? Use ordinal encoding.. 
#exercise for you

df["ExerciseAngina"] = ord_enc.fit_transform(df[["ExerciseAngina"]])
```

```python
df.head()
```

```python
X = df.to_numpy() # Put the features in X. It has to be an array (not dataframe)
X
```

```python
#Split the data into training and testing.

from sklearn.model_selection import train_test_split

[X_train,X_test,y_train,y_test] = train_test_split(X, y, test_size= 0.2)
```

```python
#Exercise for you:

#import
from sklearn.tree import DecisionTreeClassifier

#Build a DT classifier
clf = DecisionTreeClassifier()

#Train it using the training set
clf.fit(X_train, y_train)

#Test it using the testing set and find the accuracy score
print(clf.score(X_test,y_test))

#PLot the DT
```

```python
y_predicted = clf.predict(X_test)
```
