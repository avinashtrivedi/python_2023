```python
import os
import pandas as pd

# change here
# main folder path containing march 1,2,3 folder
path = r'D:\OneDrive - NITT\CODE\Python Scripts\Code_2023\Folder'
```

```python
data = []
for DateFolder in os.listdir(path):
    temp_path = os.path.join(path,DateFolder)
    excel_fpath = os.path.join(temp_path,'Fails Report.xlsx')
    
    # change sheet name here
    df = pd.read_excel(excel_fpath,sheet_name='Data')
    data.append(df)
data = pd.concat(data)
data.reset_index(drop=True,inplace=True)
data_no_duplicate = data.drop_duplicates(subset='TradeRef')
data_no_duplicate.reset_index(drop=True,inplace=True)
data_no_duplicate.sort_values('ops Ageing ',ascending=False)
```

```python
writer = pd.ExcelWriter('Merged_excel.xlsx', engine='xlsxwriter')
data.to_excel(writer, sheet_name='final_appended_data',index=False)
data_no_duplicate.to_excel(writer, sheet_name='without_duplicate&sorted',index=False)
writer.close()
```
