```python
import pandas as pd
```

```python
df = pd.read_excel('D:\OneDrive - NITT\Custom_Download\RAC.xlsx')
```

```python
df
```

```python
x = df.groupby('Due Date Bucket')['Due Date Bucket'].count()
```

```python
x = pd.DataFrame(x)
```

```python
x.columns = ['count']
```

```python
x = x.reset_index()
```

```python
x['percent'] = 1/x['count']*100
```

```python
x
```

```python
x.drop('count',axis=1,inplace=True)
```

```python
d = {}
for indx,line in x.iterrows():
    k = line['Due Date Bucket']
    v = line['percent']
    d[k] = v
```

```python
df['percent'] = df['Due Date Bucket'].apply(lambda x:d[x])
```

```python
df
```

```python
df
```

```python
df['Quarter'] = df['Due Date RAV'].apply(lambda x:(x.month-1)//3)
```

```python
df
```

```python
df['Due Date RAV'] = df['Due Date RAV'].fillna(df['Due Date RAV'].mode()[0])
```

```python
import matplotlib.pyplot as plt
```

```python
y = df[['Due Date Bucket','percent']].drop_duplicates()
```

```python
y
```

```python
plt.bar(y['Due Date Bucket'],y['percent'])
plt.xticks(rotation=90)
plt.show()
```

```python
month = {5:'May',6:'June',7:'July'} # make it 1 to 12
```

```python
df['month'] = df['Due Date RAV'].apply(lambda x:month[int(x.month)])
```

```python
df['Year'] = df['Due Date RAV'].apply(lambda x:int(x.year))
```
