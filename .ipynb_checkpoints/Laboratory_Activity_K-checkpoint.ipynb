```python
df_hspt = df_hspt.rename(columns = {"Facility ID": "FIPS"})
```

```python
df_hspt['FIPS'] = df_hspt['FIPS'].astype('int')
```

```python
df_pple = dt['People']
```

```python
df = pd.merge(df_hspt,df_pple,  on='FIPS')
```

```python
Y = df_hspt.groupby('County Name')['Hospital Type'].count()
```

```python
df_pple['County'] = df_pple['County'].apply(lambda x:x.upper())
```

```python
X = df_pple[['FIPS','County','TotalPopEst2019']]
```

```python
Y = Y.reset_index()
```

```python
Y = Y.rename(columns = {"County Name": "County"})
```

```python
X_copy = X.copy()
```

```python
X = X.groupby('County')['TotalPopEst2019'].sum().reset_index()
```

```python
df = pd.merge(X,Y,  on='County')
```

```python
df['people_per_psychiatric'] = df['TotalPopEst2019']/df['Hospital Type']
```

```python
X_copy[['FIPS','County']]
```

```python
df = pd.merge(X_copy[['FIPS','County']],df,  on='County')
```

```python
df_final = df[['FIPS','people_per_psychiatric']]
```

Creating choropleth for the variable that has to be done above

```python
dt = dt.dropna() # first we have to clean our data by dropping NA values if there are any
dt
```

```python
dt.reset_index() #removing gaps from index column if there ar any
```

```python
from urllib.request import urlopen
import json
with urlopen('https://raw.githubusercontent.com/plotly/datasets/master/geojson-counties-fips.json') as response:
    counties = json.load(response)
```

```python
counties["features"][10]
```

```python
import plotly.express as px

fig = px.choropleth_mapbox(df_final, geojson=counties, locations='FIPS', color='people_per_psychiatric',
                           color_continuous_scale="Viridis",
                           range_color=(0, 12),
                           mapbox_style="carto-positron",
                           zoom=3, center = {"lat": 37.0902, "lon": -95.7129},
                           opacity=0.5,
                           labels={'people_per_psychiatric':'number of people per psychiatric hospitals in the counties'}
                          )
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```

Now this is the actual code for the chorolopleth using the ploty package

```python
fig = px.choropleth_mapbox(dt, geojson=counties, locations='FIPS', color='PctEmpChange1819',
                           color_continuous_scale="jet",
                           range_color=(-10, 10),
                           mapbox_style="open-street-map",
                           zoom=2.5, center = {"lat": 37.0902, "lon": -95.7129},
                           opacity=0.5,
                           labels={'PctEmpChange1819':'Unemployment Rate Change<br>(2018-19)'}
                          )

fig.show()
```

Centering on the #1 main place with psychiatric hospitals (here we have Galesburg but we have to change it to what applies to this activity which is not going to be Galesburg)

```python
# Centering on Galesburg 

fig = px.choropleth_mapbox(dt, geojson=counties, locations='FIPS', color='Metro2013',
                           mapbox_style="carto-positron",
                           zoom=6, opacity=0.15,
                           center = {"lat": 40.9522, "lon": -90.3686},
                           labels={'Metro2013':'Metro Area<br>(2013)'}
                          )

fig.show()
```

improving map

```python

mask = dt["Metro2013"] > 0.500
dt.loc[ mask, "Metro2013c"] = "Metro County"
dt.loc[~mask, "Metro2013c"] = "Non-Metro County"
    
```

```python
# Improving the Map

fig = px.choropleth_mapbox(dt, geojson=counties, locations='FIPS', color='Metro2013c',
                           mapbox_style="carto-positron",
                           zoom=6, opacity=0.5,
                           center = {"lat": 40.9522, "lon": -90.3686},
                           labels={'Metro2013c':'Area Classification'},
                           color_discrete_map={'Metro County':'#880000','Non-Metro County':'#00aa00'},
                           hover_name='County_x'
                          )

fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})

fig.show()
```

```python
mask = dt["PersistentChildPoverty2004"] > 0.500
dt.loc[ mask,"PersistentChildPoverty2004pn"] = "Persistent"
dt.loc[ ~mask,"PersistentChildPoverty2004pn"] = "Not Persistent"


dt.loc[ pd.isna(dt["PersistentChildPoverty2004"]), "PersistentChildPoverty2004pn"] = "Not Recorded"
```

```python
fig = px.choropleth_mapbox(dt, geojson=counties, locations='FIPS', color='PersistentChildPoverty2004pn',
                           mapbox_style="carto-positron",
                           zoom=3, center = {"lat": 37.0902, "lon": -95.7129},
                           opacity=1,
                           labels={'PersistentChildPoverty2004pn':'Child Poverty (2004)'},
                           color_discrete_map={'Persistent':'red','Not Persistent':'#008800','Not Recorded':'blue'},
                           hover_name='County_x',
                           category_orders={"PersistentChildPoverty2004pn": ["Persistent","Not Persistent","Not Recorded"]},
                          )

fig.update_traces(marker_line_width=0.5)
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})

fig.show()
```
