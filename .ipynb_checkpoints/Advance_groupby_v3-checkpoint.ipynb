```python
def fun(d1,d2):
    t = {x:min(d1[x],d2[x]) for x in d1 if x in d2}
    return t
```

```python
def non_match_fun(d1,d2):
    all_key = {**d1,**d2}
    key_del = set(d1).intersection(set(d2))
    return {i:all_key[i] for i in all_key if i not in key_del}
```

```python
def custom_agg(x):
    return pd.Series({
        
    'unique_BUY_cur': x.BUY_CCY_ID.nunique(), 
    'count_BUY_cur': dict(x.BUY_CCY_ID.value_counts()),
    'unique_SELL_cur': x.SELL_CCY_ID.nunique(),
    'count_SELL_cur': dict(x.SELL_CCY_ID.value_counts()),
    'diff':np.average(x.BUY_AMOUNT - x.SELL_AMOUNT),
    'Matching_curr': fun(dict(x.BUY_CCY_ID.value_counts()),dict(x.SELL_CCY_ID.value_counts())),
    'Matching_curr_count': sum(fun(dict(x.BUY_CCY_ID.value_counts()),dict(x.SELL_CCY_ID.value_counts())).values()),
    'Non Matching_curr': non_match_fun(dict(x.BUY_CCY_ID.value_counts()),dict(x.SELL_CCY_ID.value_counts())),
    'Non Matching Count': sum(non_match_fun(dict(x.BUY_CCY_ID.value_counts()),dict(x.SELL_CCY_ID.value_counts())).values()),
        
    })
```

```python
x = df.groupby(['COMPANY_ID','LEGAL_ENTITY_ID','VALUE_DATE']).apply(custom_agg).reset_index()
```
