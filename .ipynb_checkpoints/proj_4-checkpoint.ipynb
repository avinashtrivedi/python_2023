# Q1

```python
# import necessary libraries
import pandas as pd
import numpy as np
from tqdm import tqdm
import matplotlib.pyplot as plt
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix, f1_score,ConfusionMatrixDisplay
```

```python
# read the dataset
df = pd.read_csv(r"D:\OneDrive - NITT\Custom_Download\Assignment4_Q1.csv")
```

```python
# check for categorical columns
df.select_dtypes('object').columns
```

```python
# convert categorical columns to one-hot encoding
df = pd.get_dummies(df)
```

```python
# display top 5 rows
df.head()
```

```python
# get X and y
X = df.drop('Attrition',axis=1)
y = df['Attrition']
```

```python
# split the data into training and validation set in 60:40 ratio
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.40, random_state=42)
```

```python
# build Naive Bayes classifier
gnb = GaussianNB()

# train  naive Bayes classifier on training set
gnb.fit(X_train, y_train)

# perform prediction on validation set
y_pred = gnb.predict(X_val)
```

```python
def get_result(y_true,y_pred):
    
    # get the confusion matrix
    tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()
    
    # compute Specificity,Sensitivity,Accuracy and f1
    TNR = tn/(tn+fp) 
    TPR = tp/(tp+fn)
    ACC = (tp+tn)/(tp+fp+fn+tn)
    f1 = f1_score(y_true, y_pred)

    print("Specificity:",round(TNR,3))
    print("Sensitivity:",round(TPR,3))
    print("Accuracy:",round(ACC,3))
    print("F1:",round(f1,3))
    
    print('---------Confusion Matrix--------------')
    cm = confusion_matrix(y_true, y_pred, labels=[0,1])
    disp = ConfusionMatrixDisplay(confusion_matrix=cm,
                                  display_labels=[0,1])
    # display confusion matrix
    disp.plot()

    plt.show()
```

```python
# print model performance
get_result(y_val,y_pred)
```

- Considering the size of the given dataset, Model is working good with accuracy of 73%

# Q2

```python
# load the dataset
df = pd.read_csv(r"D:\OneDrive - NITT\Custom_Download\ToyotaCorolla_Assignment_4_Q3.csv")
```

```python
# check the data types
df.dtypes
```

```python
# save features to be used in the task as per the given question
features = ['Age_08_04', 'KM', 'Fuel_Type', 'HP', 'Automatic', 'Doors', 'Quarterly_Tax', 'Mfr_Guarantee', 'Guarantee_Period', 'Airco',
'Automatic_airco', 'CD_Player', 'Powered_Windows', 'Sport_Model', 'Tow_Bar']
```

```python
# perform one-hot encoding for categorical columns and get X and y
X = pd.get_dummies(df[features])
y = df['Price']
```

```python
X
```

```python
y
```

```python
# split the data into training and validation set in 60:40 ratio
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.40, random_state=42)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_val = sc_X.transform(X_val)
```

```python
# import Decision Tree Regressor
from sklearn.tree import DecisionTreeRegressor
```

```python
# Build the DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state=42)
```

```python
# train the model on training set
regressor.fit(X_train,y_train)
```

```python
# perform prediction on validation set
y_pred = regressor.predict(X_val)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error,r2_score

# function to print regression result
def get_reg_result(y_true,y_pred):
    
    # mean squared error
    mse = mean_squared_error(y_true,y_pred,squared=True)
    
    # Root mean squared error
    rmse = mean_squared_error(y_true,y_pred,squared=False)
    
    # mean absolute error 
    mae = mean_absolute_error(y_true,y_pred)
    
    # R2
    r2 = r2_score(y_true,y_pred)
    
    print('MSE:',mse)
    print('RMSE:',rmse)
    print('MAE:',mae)
    print('R2:',r2)
```

```python
# print regression model result
get_reg_result(y_val,y_pred)
```
