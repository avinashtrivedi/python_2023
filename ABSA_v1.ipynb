```python
import re
import unicodedata as uni

def remove_url(text):
    text = re.sub(r"http\S+", "", text)
    return text

import demoji


def handle_emoji(string):
    emojis = demoji.findall(string)

    for emoji in emojis:
        string = string.replace(emoji, " " + emojis[emoji].split(":")[0])

    return string

def word_tokenizer(text):
    text = text.lower()
    text = text.split()

    return text

import spacy

sp = spacy.load("en_core_web_sm")

def lemmatization(text):

    text = " ".join(text)
    token = sp(text)

    text = [word.lemma_ for word in token]
    return text

from nltk.corpus import stopwords

en_stopwords = set(stopwords.words('english'))

def remove_stopwords(text):
    text = [word for word in text if word not in en_stopwords]
    return text


def preprocessing(text):
    
    text = remove_url(text) 
    text = uni.normalize('NFKD', text)
    text = handle_emoji(text)
    text = text.lower() 
    text = re.sub(r'[^\w\s]', '', text)
    text = word_tokenizer(text)
    text = lemmatization(text)
    text = remove_stopwords(text)
    text = " ".join(text)

    return text
```

```python
df['clean_review'] = df['review'].progress_map(preprocessing)
```

```python
data_words = df['clean_review1'].values.tolist()
len(data_words)
```

```python
import gensim.corpora as corpora

id2word = corpora.Dictionary(data_words)

texts = data_words

corpus = [id2word.doc2bow(text) for text in texts]
```

```python
from gensim.models import LdaMulticore
from gensim.models import LdaModel
from pprint import pprint

num_topics = 5
# Build LDA model
lda_model = LdaMulticore(corpus=corpus, id2word=id2word,
                     num_topics=num_topics, iterations=400)


pprint(lda_model.print_topics())
doc_lda = lda_model[corpus]
```

```python
# Phone
# Camera
# Battery
# Delivery
# Processor
```

```python
from gensim.models import FastText
fasttext_model = FastText(data_words, vector_size= 100, window=1, min_count=1, workers=4,sg=1)
```

```python
fasttext_model.wv.n_similarity("I really like the camera of this phone", "battery")
```

```python
aspects = ["phone", "camera", "battery", "delivery", "processor"]

def get_similarity(text, aspect):
    try:
        text = " ".join(text)
        return fasttext_model.wv.n_similarity(text, aspect)
    except:
        return 0
```

```python
for aspect in aspects:
    df[aspect] = df['clean_review1'].progress_map(lambda text: get_similarity(text, aspect))
```

```python
df.to_csv("ABSA_final.csv", index = False)
```
