```python
def merge(df_snow,df_emap):
    '''
    Make sure to enter the respective dataframe in the same sequence.
    
    df_snow: Servicenow Dataframe
    df_emap:Emapps Dataframe
    
    return: merged dataframe based on id,ref and inc
    '''
    df = pd.DataFrame([],columns=list(df_snow)+['X','Y'])
    for indx, line in df_snow.iterrows():
        x,y = None,None
        snow_id = line['Problem_DBIRS_EVENT_ID']
        if snow_id in df_emap['EMAPP_Event ID'].values:
            x,y = df_emap[df_emap['EMAPP_Event ID']==snow_id][['X','Y']].values.tolist()[0]
        else:
            ref_num = line['Incident Reference Number']
            temp = df_emap[df_emap['Emapp_Reference ID'].apply(lambda t: True if ref_num in t.split() else False)]
            if len(temp):
                x,y = temp[['X','Y']].values.tolist()[0]
            else:
                inc_num = line['Incident Technology Incident Ref']
                other_temp = df_emap[df_emap['Emapp_Reference ID'].apply(lambda t: True if inc_num in t.split() else False)]
                if len(other_temp):
                    x,y = other_temp[['X','Y']].values.tolist()[0]

        if x and y:
            df.loc[len(df)] = list(line) + [x,y]
            
    df = df[['Problem_DBIRS_EVENT_ID','Incident Reference Number','P','Q','X','Y','Incident Technology Incident Ref']]
    return df
```

```python
merge(df_snow,df_emap)
```
