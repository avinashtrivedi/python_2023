## A-5

```python
# Answers to A-5 

null_columns=df.columns[df.isna().any()]
print('Columns with NaN values are:', null_columns)

#Resolving the inconsistency in Age column
df["Age"]=df["Age"].apply(lambda x: int(x.replace("yrs","")))
 
#Resolving the inconsistency in HS column
df["HS Score"]=df["HS Score"].apply(lambda x: int(x.replace("%","")))
display(df.info())

#check if there is negative values
seleted_cols= df.select_dtypes(exclude='object').columns
df[seleted_cols]=df[seleted_cols].applymap(lambda x: x if x>=0 else abs(x))
display(df) 
```

## A-6

```python
# Answers to A-6
print(df.select_dtypes("object").describe().loc['unique', :])
from sklearn.preprocessing import LabelEncoder

encoder1 = LabelEncoder()
encoder1.fit(df['Gender'])
df['Gender']=encoder1.transform(df['Gender'])

#Siblinges column, yes=1, No=1
encoder2 = LabelEncoder()
encoder2.fit(df['Siblings'])
df['Siblings']=encoder2.transform(df['Siblings'])
                
#Admission Status column, yes=1, No=0
encoder3 = LabelEncoder()
encoder3.fit(df['Admission Status'])
df['Admission Status']=encoder3.transform(df['Admission Status'])

#One hot encoding on column Parents Edu. Remove the original column.
df=pd.get_dummies(df, columns=['Parents Edu.'],drop_first=True)

#Location column
Lo_mapping = {'Western':1, 
               'Central':2, 
               'Eastern':3}
df['Location'] = df['Location'].map(Lo_mapping)
                
#School Rank
Rank_mapping = {'Low':1, 
               'Avg':2, 
               'High':3}
df['School Rank'] = df['School Rank'].map(Rank_mapping)

display(df.head())
df.to_csv('HW2Data_A_Clean.csv', index = True, header=True)
```

## A-8

```python
# Answers to A-8
#exclude ordinal columns
num_columns = df.select_dtypes(exclude='object').columns
#exclude output column
pca_input_columns=num_columns.drop(['Admission Status'])

#Find the first two PCs
from sklearn.decomposition import PCA
pca = PCA(n_components=3)
principalComponents = pca.fit_transform(df[pca_input_columns])
df['pca1'] = principalComponents[:,0]
df['pca2'] = principalComponents[:,1]

#variation captured by each of the principal components.
for ind,var in enumerate(pca.explained_variance_ratio_):
    print(f'The variance captured by pca {ind+1} is: {var*100: 0.2f}%')
    
#the coefficients of the linear combination of the columns and the labels of three of three columns having the highest coefficients.
import numpy as np
 
pca_1_coefficients=np.round(pca.components_[0],4)
zipList=list(zip(pca_input_columns,pca_1_coefficients))
ndf=pd.DataFrame.from_records(zipList,columns=['field','coeff'])
display(ndf.sort_values(by=['coeff'], ascending=False))


#scatter plot using first two PCs differentiate by Admission Status
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure()
sns.scatterplot(x='pca1',y='pca2',
            hue='Admission Status', palette=['r','b','g'],
            alpha=0.75,
            data=df)
plt.title('PCA vs Admission Status', fontsize = 20)
plt.show()
```

## B-3

```python
# Answers to B-3

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Lasso

import warnings
warnings.filterwarnings('ignore')
```

```python
# Answers to B-3
 #B3

df = pd.read_csv("HW2DataB.csv")
df_c = pd.read_csv("HW2DataC.csv")

# Remove the "Student ID" column permanently
df = df.drop("StdID", axis=1)

# Data transformation using Min-Max normalization on numeric columns (columns 5 to 11)
numeric_columns = df.columns[5:]
df[numeric_columns] = (df[numeric_columns] - df[numeric_columns].min()) / (df[numeric_columns].max() - df[numeric_columns].min())

# Display the summaries of all columns in transpose form
summary_transpose = df.describe().T

print(summary_transpose)
```

## B-4

```python
# Answers to B-4
 


# Relation plot using pandas.plotting.scatter_matrix
sns.pairplot(df[numeric_columns])
plt.suptitle('Relation Plot')
plt.show()


```

## B-5

```python
#B5

# Define the explanatory variables (X) and the explained variable (Y)
X = df[numeric_columns].values
Y = df['Final'].values

# Add a column of ones to X for the intercept term
X_with_intercept = np.column_stack((np.ones(X.shape[0]), X))

# Calculate the OLS coefficient estimates using the formula ðœƒ = (ð‘‹ð‘‡ð‘‹)âˆ’1 ð‘‹ð‘‡ð‘Œ
theta = np.linalg.inv(X_with_intercept.T.dot(X_with_intercept)).dot(X_with_intercept.T).dot(Y)

# Extract the intercept and slope values
intercept_ols = theta[0]
slopes_ols = theta[1:]

# Print the obtained intercept and slope values
print("Intercept:", intercept_ols)
print("Slopes:", slopes_ols)

# Print the final regression model
print("Final Regression Model:")
print("Final =", f"{intercept_ols:.2f}", end=" ")
for i, column in enumerate(numeric_columns):
    print(f"+ {slopes_ols[i]:.2f} * {column}", end=" ")
print()
```

## B-6

```python
# Answers to B-6
 #B6

# Calculate the OLS coefficient estimates using sklearn's LinearRegression
model = LinearRegression()
model.fit(X, Y)

# Extract the intercept and slope values
intercept_lr = model.intercept_
slopes_lr = model.coef_

# Print the OLS coefficients obtained from sklearn
print("Intercept (sklearn):", intercept_lr)
print("Slopes (sklearn):", slopes_lr)

# Compare the coefficients obtained from Part B-5 with those from sklearn
difference_in_intercept = intercept_ols - intercept_lr
difference_in_slopes = slopes_ols - slopes_lr

print("\nDifferences in Coefficients:")
print("Intercept:", difference_in_intercept)
print("Slopes:", difference_in_slopes)

# Find variables with insignificant regression coefficients
insignificant_variables = []
for i, column in enumerate(numeric_columns):
    if abs(difference_in_slopes[i]) < 1e-5:  # Adjust the threshold as needed
        insignificant_variables.append(column)

print("\nVariables with Insignificant Coefficients:", insignificant_variables)

```

## B-7

```python
# Answers to B-7
# Answers to B-7
#B7

# Remove the "Student ID" column permanently
df_c = df_c.drop("StdID", axis=1)

# Data transformation using Min-Max normalization on numeric columns (columns 5 to 11)
numeric_columns_c = df_c.columns[5:]
df_c[numeric_columns_c] = (df_c[numeric_columns_c] - df_c[numeric_columns_c].min()) / (df_c[numeric_columns_c].max() - df_c[numeric_columns_c].min())

# Define the explanatory variables (X) and the explained variable (Y) for HW2DataC
X_c = df_c[numeric_columns_c]
Y_c = df_c['Final']

# Use the OLS coefficient estimates obtained from B-6
model = LinearRegression()
model.intercept_ = intercept_lr
model.coef_ = slopes_lr

# Calculate the predicted values for HW2DataC using the OLS model
predicted_Y_c = model.predict(X_c)

# Calculate the Mean Squared Error (MSE) for HW2DataC
mse_c = mean_squared_error(Y_c, predicted_Y_c)

print("Mean Squared Error (MSE) for HW2DataC:", mse_c)
 


 
```

## B-8

```python
# Answers to B-8


# Define the explanatory variables (X) and the explained variable (Y) for HW2DataB
X_b = X
Y_b = Y

# Define the explanatory variables (X) and the explained variable (Y) for HW2DataC
X_c = df_c[numeric_columns_c]
Y_c = df_c['Final']

# Perform Ridge Regression with cross-validation to find the best alpha value
alphas = [1e-3, 1e-2, 1e-1, 1e0, 1e1, 1e2, 1e3]
ridge = Ridge()
param_grid = {'alpha': alphas}
grid_search = GridSearchCV(ridge, param_grid, cv=7)
grid_search.fit(X_b, Y_b)
best_alpha_ridge = grid_search.best_params_['alpha']

# Calculate the predicted values for HW2DataC using Ridge Regression with the best alpha value
ridge_model = Ridge(alpha=best_alpha_ridge)
ridge_model.fit(X_b, Y_b)
predicted_Y_c = ridge_model.predict(X_c)

intercept_ridge = ridge_model.intercept_
slopes_ridge = ridge_model.coef_

# Calculate the Mean Squared Error (MSE) for HW2DataC using the Ridge model
mse_c = mean_squared_error(Y_c, predicted_Y_c)

print("Best Alpha:", best_alpha_ridge)
print("Mean Squared Error (MSE) for HW2DataC using Ridge Regression:", mse_c) 


```
